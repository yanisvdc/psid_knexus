# gaussican copula

Our data dictionary (first_dict) may be inconsistent with the metadata dictionary inferred by the sdv method (second_dict).

sdv metadata is \\
{columns: {  age_87: {sdtype: numerical}  }, {  age_88: {sdtype: numerical}  } ... }

feature_type_dict is \\
{categorical: list, ordinal: list, continuous: list}

To compare them, for each, I create a dictionary of the form {feature name: type, ...}, and print feature name when the types do not coincide, I replace continuous by numerical in the first_dict

```python
from sdv.metadata import SingleTableMetadata
import json
```

```python
import ijson

def my_dict(name):
    # Path to the JSON file (assuming it's in the same folder as your script)
    json_file_path = paths[name]["dict"]

    # Create the dictionary with keys 'categorical', 'ordinal', and 'continuous'
    feature_type_dict = {
        "categorical": [],
        "ordinal": [],
        "continuous": []
    }

    # Stream through the JSON data incrementally
    with open(json_file_path, 'r') as file:
        parser = ijson.items(file, '')

        # Extract the first object, which contains the JSON key-value pairs
        json_data = next(parser)

    # Parse through the JSON structure and assign column names to the corresponding feature types
    for column, attributes in json_data.items():
        feature_type = attributes.get("feature type")

        if feature_type in feature_type_dict:
            feature_type_dict[feature_type].append(column)
        else:
            print(f"Unknown feature type for column '{column}': {feature_type}")

    # remove year, and merge ordinal and categorical as ordinal is not a recognized type by sdv synthesizers
    feature_type_dict["ordinal"].remove("year")
    feature_type_dict['categorical'].extend(feature_type_dict['ordinal'])
    del feature_type_dict['ordinal']

    # PB: the feature_type dictionary does not contain the "_year" features at this point

    # Process each feature type
    for data_type in feature_type_dict:
        updated_features = []

        for feature in feature_type_dict[data_type]:
            if feature in paths[name]["constant"]:
                updated_features.append(feature)
            else:
                updated_features.extend([f"{feature}_{i}" for i in range(87, 97)])

        # Replace the old list with the updated one
        feature_type_dict[data_type] = updated_features

    return feature_type_dict
```

```python
def preproc_my_dict(first_dict):
    new_first_dict = {
      "numerical" if key == "continuous" else key: value
      for key, value in first_dict.items()
      }

    # Reverse first_dict to map feature names to types
    reversed_first_dict = {}
    for key, values in new_first_dict.items():
        for value in values:
            reversed_first_dict[value] = key

    return reversed_first_dict
```

```python
# Function to find and print discrepancies
def check_discrepancies(reversed_first_dict, second_dict):
    '''
    reversed_first_dict: my reversed dictionary {feature name: type}
    second_dict: sdv metadata dictionary
    '''
    print("Checking for discrepancies...")

    columns_with_specific_types = {}

    columns_with_wrong_types = []

    # Iterate through the second_dict and compare
    for col_name, col_info in second_dict["columns"].items():
        # col_name is the feature name, col_info is a dictionary e.g. {'sdtype': "numerical"}
        second_type = col_info["sdtype"]
        first_type = reversed_first_dict.get(col_name)

        if second_type not in ["categorical", "numerical"]:
            columns_with_specific_types[col_name] = second_type
        else:
            # second_type is either categorical or numerical but differs from reversed_first_dict
            # enforce the reversed_first_dict type
            if first_type != second_type:
                columns_with_wrong_types.append(col_name)

    print(f"Columns with specific types:{columns_with_specific_types}")
    print(f"Fail if > 0: Number of columns with wrong types: {len(columns_with_wrong_types)}")
    print("Done")
    return columns_with_wrong_types, columns_with_specific_types
```

```python
from sdv.single_table import GaussianCopulaSynthesizer

def gaussian_copula(name):
    df = pd.read_csv(paths[name]['data'])

    metadata = SingleTableMetadata()
    metadata.detect_from_dataframe(df)

    feature_type_dict = my_dict(name)
    reversed_first_dict = preproc_my_dict(feature_type_dict)

    # Check discrepancies and return list of columns where the reversed_first_dict type needs to be enforced
    # i.e. wrong type is inferred by sdv metadata
    print("First check")
    columns_with_wrong_types, _ = check_discrepancies(reversed_first_dict, metadata.to_dict())
    for col in columns_with_wrong_types:
        metadata.update_column(
            column_name=col,
            sdtype=reversed_first_dict[col]
        )
    print("Second check after update")
    _, __ = check_discrepancies(reversed_first_dict, metadata.to_dict())

    metadata.validate()
    synthesizer = GaussianCopulaSynthesizer(metadata)
    synthesizer.fit(df)

    synthetic_data = synthesizer.sample(num_rows=df.shape[0])

    return synthetic_data
```
